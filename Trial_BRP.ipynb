```python
import librosa
x, sr = librosa.load('Harry.mp3', sr=None)
```

```python
print(x.shape)
print(sr)
dur = librosa.core.get_duration(x,sr)
print(dur)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14,5))
librosa.display.waveplot(x, sr=sr)
plt.title('Waveform of the Audio File')
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.title('Spectogram')
```

```python
import os
import librosa
import fnmatch



counter=1

classes = []
duration =[]
instruments = ['banjo', 'bass-clarinet', 'bassoon', 'cello', 'clarinet', 'contrabassoon', 'english-horn', 'double bass', 'flute', 'french-horn', 'guitar', 'mandolin', 'oboe', 'percussion', 'saxophone', 'trombone', 'trumpet', 'tuba', 'viola', 'violin']
    
    
for root,dirnames,filenames in os.walk('/Users/DELL/Notebookks/all-samples/'):
    for i,filename in enumerate(filenames):
        print("Get %d = %s" %(counter,filename))
        counter+=1
        if 'percussion' in root:
            classes.append('percussion')
        else:
                for instrument in instruments:
                    if fnmatch.fnmatchcase(filename, instrument+'*'):
                        classes.append(instrument)
import pandas as pd
ds_classes = pd.Series(classes)
a = ds_classes.value_counts()
print(a)                        


 

        


```
